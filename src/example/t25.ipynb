(**Identity Matrix are orthogonal [$v \times w = 0$]**)

- Given Identity Matrix ($I_{3}$):

$$
\begin{bmatrix}
1 & 0 & 0 \\\\
0 & 1 & 0 \\\\
0 & 0 & 1
\end{bmatrix}
$$

1. With paper and pencil, use the dot product to demonstrate that any 2 columns of this $I_{3}$ matrix are orthogoa=nal to each other. (Code)

```python
import numpy as np

i = np.array([[1,0,0], [0,1,0], [0,0,1]])

for idx in range(len(i)-1):
    orthogonal: bool = True if (np.dot((i[idx]), (i[idx+1])) == 0) else False
    print(orthogonal)
```

```python
import numpy as np

i = np.array([[1,0,0], [0,1,0], [0,0,1]])

for _ in range(len(i)):
    print(i[_])
```
